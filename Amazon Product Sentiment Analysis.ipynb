```python
# importing necessary libraries
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns # visualization
import matplotlib.pyplot as plt
```

# Data Collection

```python
# loading the data
df=pd.read_csv("Amazon-Product-Reviews - Amazon Product Review (1).csv")
```

```python
df.head(5)
```

# Data Preprocessing

```python
# checking the data info
df.info()
```

```python
# checking the number of rows and columns
df.shape
```

```python
# checking the count of null 
df.isna().sum()
```

```python
# droping the null records
df.dropna(inplace=True)
```

```python
# checking the count of null again
df.isna().sum()
```

```python
# converting from Dataframe to datetime 
df['review_date'] = pd.to_datetime(df['review_date'])
```

```python
df['review_year'] = df['review_date'].dt.year
df['review_month'] = df['review_date'].dt.month
```

```python
df.head(5)
```

```python
'''
The code snippet calculates the mean star rating for each group, and then resets the index to create a flat DataFrame with columns for 'review_year', 'review_month', 
and the corresponding mean star rating.
'''
star_rating_mean_values = df.groupby(['review_year',"review_month"])['star_rating'].mean().reset_index()
star_rating_mean_values
```

# Data Visualization

```python
# visualizing the graph
sns.lineplot(data=star_rating_mean_values,x="review_month",y="star_rating",hue="review_year")
```

The observed trend suggests that customers may tend to leave more positive reviews during winter months (late year and early next year) compared to spring and summer months, where ratings may slightly decrease. This seasonal variation in average star ratings could be influenced by factors such as holiday seasons, weather changes, or other seasonal patterns affecting customer experiences and sentiment.

```python
# calculates the average star rating ("star_rating") for each sentiment category - positive and negative
star_rating_mean_values_by_sentiment=df.groupby("sentiment",)["star_rating"].mean().reset_index()
```

```python
star_rating_mean_values_by_sentiment
```

```python
sns.barplot(data=star_rating_mean_values_by_sentiment,x="sentiment",y="star_rating")
```

Average rating for negative sentiment is 2.1 and Average rating for positive sentiment is 4.77

```python
# It groups the data by these columns, computes the mean star rating within each group
star_rating_mean_values_by_sentiment=df.groupby(["sentiment",'review_year',"review_month"])["star_rating"].mean().reset_index()
```

```python
star_rating_mean_values_by_sentiment
```

```python
# Create the bar plot
plt.figure(figsize=(12, 6))
ax = sns.barplot(data=star_rating_mean_values_by_sentiment, x="review_year", y="star_rating", hue="sentiment")

# Add labels to the bars
for container in ax.containers:
    ax.bar_label(container, fontsize=10)

# Set title and labels
plt.title('Mean Star Rating by Sentiment and Review Year')
plt.xlabel('Review Year')
plt.ylabel('Mean Star Rating')

# Show the plot
plt.show()
```

The data suggests distinct patterns in average star ratings based on sentiment categories (0 for negative, 1 for positive) over time (by month and year).
Positive sentiment reviews consistently receive higher ratings compared to negative sentiment reviews.
This analysis provides insights into the relationship between sentiment and average star ratings, highlighting trends in customer sentiment across different periods.


# Conclusion

Based on these insights, it can be concluded that seasonal factors may play a role in influencing customer sentiment and ratings, with higher satisfaction levels observed during winter months. Additionally, there is a clear distinction in average ratings between positive and negative sentiment reviews, emphasizing the importance of customer sentiment in determining overall ratings. This analysis underscores the need for businesses to monitor and understand customer sentiment trends to enhance customer experiences and satisfaction levels throughout different periods of the year.
